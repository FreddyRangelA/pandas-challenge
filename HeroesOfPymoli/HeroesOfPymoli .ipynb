```python
#dependencies
import pandas as pd
```

```python
#make a references to the Heroes files
csv_path="Resources/purchase_data.csv"

#read the file and return the table
heroes_df=pd.read_csv(csv_path)
heroes_df = heroes_df.dropna(how='any')
heroes_df

heroes_df.head()
```

```python
#get the number of unique items on the SN column 
count=heroes_df["SN"].nunique()

#get the number of unique items on the item name column 
items=heroes_df["Item Name"].nunique()

numberOfPurchases=heroes_df["Purchase ID"].count()

revenue=heroes_df["Price"].sum()


#print(revenue)
#print(numberOfPurchases)
#print(items)
#print(count)
```

```python
# Player Count
#Create the data frame for total players. 
player_df=pd.DataFrame({"Total payer":[count]})
player_df
```

```python
# Purchasing Analysis (Total)

#returns the average price
averagePrice=heroes_df["Price"].mean()

#creats data frame for averga price
purchasing_df=pd.DataFrame({"Number of unique items":[items],
                            "average Price":[averagePrice],
                            "Purchase ID":[numberOfPurchases],
                           "Price":[revenue]})
purchasing_df
```

```python
#Gender Demographics
gender=heroes_df.groupby("Gender").nunique()
gender_df=gender[["SN"]]

gender_df.head()
```

```python
#df[percent] = (df['column_name'] / df['column_name'].sum()) * 100
perc_df= pd.DataFrame(gender_df)
perc_df['percent'] = round(((gender_df['SN']/gender_df['SN'].sum())*100),2)
perc_df
```

```python
purchases_analisysG = heroes_df.groupby("Gender").count()
purchases_analisysG_df=purchases_analisysG[["Purchase ID"]]
purchases_analisysP = heroes_df.groupby("Gender").sum()
purchases_analisysP_df=purchases_analisysP[["Price"]]
purchases_analisys_df=pd.merge(purchases_analisysG_df,purchases_analisysP_df,how='outer',on='Gender')

purchases_analisys_df=purchases_analisys_df.rename(columns={'Price':'Total Purchase Value'})
purchases_analisys_df=purchases_analisys_df.rename(columns={'Purchase ID':'Purchase count'})
purchases_analisys_df["Average Purchase Price"]=round(purchases_analisys_df['Total Purchase Value']/purchases_analisys_df['Purchase count'],2)
purchases_analisys_df["Avg Total Purchase per Person"]=round(purchases_analisys_df["Total Purchase Value"]/gender_df["SN"],2)
#neworder = ['EmployeeID','FirstName','MiddleName','LastName','Contact']
#df=df.reindex(columns=neworder)
neworder=['Purchase count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']
purchases_analisys_df=purchases_analisys_df.reindex(columns=neworder)
purchases_analisys_df
```

```python
#bings and lables
bing=     [0,9,14,19,24,29,34,39,45]
age_label=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]



#label_group_df=pd.DataFrame(heroes_df)
#pd.cut(label_group_df["Age"], bing, labels=age_label)

heroes_df[" "]=pd.cut(heroes_df["Age"], bing, labels=age_label)

age_group=heroes_df.groupby(" ")
unique_age=age_group["SN"].nunique()
#print(unique_age)
percentage_age=round(((unique_age/count)*100),2)
#print(percentage_age)

age_df=pd.DataFrame({"Total Count":unique_age,"Percentage of Players":percentage_age})

age_df
```

```python
#Purchasing Analysis (Age)
age_group=heroes_df.groupby(" ")
#unique values for each purchase on age rage.
ageP=age_group["Purchase ID"].nunique()

#sumation of price for each age range
averageAgeP=age_group["Price"].sum()

#averageAgeP
#ageP
average_purchase_price=averageAgeP/ageP
aPp_df=pd.DataFrame({"Purchase Count":ageP,
                     "Average Purchase Price":average_purchase_price,
                     "Total Purchase Value":averageAgeP})
aPp_df
```
